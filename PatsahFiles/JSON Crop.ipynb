```python
pt = str(3)
import json
json_data = open('./PatsahFiles/Patsah-'+pt+'-.json').read()
data = json.loads(json_data)
```

```python
import cv2
img = cv2.imread('./PatsahFiles/Patsah-'+pt+'.jpg')
plt.imshow(img)
```

```python
import csv
count = 1
for shapes in data['shapes']:
    cor1 = shapes['points'][0]
    cor2 = shapes['points'][1]
    label = shapes['label']
    crop_img=img[cor1[1]:cor2[1],cor1[0]:cor2[0]]
    direc = './PatsahFiles/Patsah-'+pt+'/'
    name = 'Patsah-'+pt+'_' + str(count) + '.jpg'
    cv2.imwrite(direc+name,crop_img)
    with open(direc+'Patsah-'+pt+'.csv','a') as csvfile:
        writer = csv.writer(csvfile,delimiter=' ')
        writer.writerow([name,label])
    print(label)
    count+=1
```
